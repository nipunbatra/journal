```python
num_clusters=100
cl = KMeans(n_clusters=num_clusters)
cl.fit(df[monthly_plus_static_features])
to_use = []
for i in range(num_clusters):
    d = cl.transform(df[monthly_plus_static_features])[:, i]
    ind = np.argsort(d)[::-1][:num_clusters]
    flag=False
    start = 0
    while flag is False: 
    
        if df[monthly_plus_static_features].index.values[ind[start]] not in to_use:
            to_use.append(df[monthly_plus_static_features].index.values[ind[start]])
            flag=True
        else:
            start = start+1
        
```
