```python
def remove_brackets(d):
    from copy import deepcopy
    e = deepcopy(d)
    for num, num_results in d.iteritems():
        for appliance_name, appliance_val in num_results.iteritems():
            if isinstance(appliance_val, dict):
                e[num][appliance_name] = np.NAN
    return e
        
```
